## **Simplifying Product Column**

```python
# First, let's see what product types actually exist in your data
# This will help us identify all the patterns we need to cover

# Check most common words in product names
from collections import Counter
import re

# Extract all words from product names
all_words = []
for name in df['product_name'].dropna():
    words = re.findall(r'\b\w+\b', name.lower())
    all_words.extend(words)

# Count most frequent words
word_counts = Counter(all_words)
print("Top 20 most common words in product names:")
for word, count in word_counts.most_common(50):
    print(f"{word}: {count}")
```

```python
# Advanced product categorization based on your actual top words
def get_product_type_advanced(name):
    name_lower = name.lower()

    # Cables & Connectivity (cable, usb, charging, type, data, sync, adapter)
    if any(word in name_lower for word in ['cable', 'usb', 'charging', 'adapter', 'sync', 'data']):
        return 'Cables & Connectivity'

    # Smart TV & Electronics (tv, smart, led, hd, remote, wireless)
    elif any(word in name_lower for word in ['tv', 'smart', 'led', 'hd', 'remote']):
        return 'TV & Smart Devices'

    # Mobile & Computing (iphone, samsung, laptop, ram, storage, pro, bluetooth)
    elif any(word in name_lower for word in ['iphone', 'samsung', 'laptop', 'ram', 'storage', 'bluetooth']):
        return 'Mobile & Computing'

    # Home Appliances (water, heater, electric, watt)
    elif any(word in name_lower for word in ['water', 'heater', 'electric', 'watt']):
        return 'Home Appliances'

    # Wearables & Accessories (watch, battery, wireless)
    elif any(word in name_lower for word in ['watch', 'battery', 'wireless']):
        return 'Wearables & Accessories'

    else:
        return 'Other Electronics'

# Apply the classification
df['product_category_new'] = df['product_name'].apply(get_product_type_advanced)

# Check results
print("Advanced product categorization:")
print(df['product_category_new'].value_counts())

# Check coverage
other_percentage = (df['product_category_new'] == 'Other Electronics').sum() / len(df) * 100
print(f"\nItems in 'Other Electronics': {other_percentage:.1f}%")
```

```python
# Even simpler approach - just truncate and clean basic stuff
def simple_clean(name):
    if pd.isna(name):
        return name

    # Remove common noise
    name = name.replace('Amazon Basics', '').replace('(', '').replace(')', '')
    name = name.replace('[', '').replace(']', '')

    # Take first part before excessive details
    if 'Compatible with' in name:
        name = name.split('Compatible with')[0]
    if ' - ' in name:
        name = name.split(' - ')[0]

    # Truncate and clean
    return name[:45].strip()

df['product_name_simple'] = df['product_name'].apply(simple_clean)

print("Simplified names:")
print(df[['product_name', 'product_name_simple']].head())
```

## **Adding Performance Column based on Rating**

```python
# Create performance categories based on rating
def categorize_performance(rating):
    if pd.isna(rating):
        return 'Unknown'
    elif rating >= 4.0:
        return 'Excellent'
    elif rating >= 3.0:
        return 'Average'
    else:
        return 'Bad'

df['performance'] = df['rating'].apply(categorize_performance)

# Check the distribution
print("Performance distribution:")
print(df['performance'].value_counts())

# Check percentage breakdown
print("\nPercentage breakdown:")
print(df['performance'].value_counts(normalize=True) * 100)
```
